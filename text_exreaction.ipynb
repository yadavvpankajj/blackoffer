```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os

# Load the Excel file
file_path = "C:\\Users\\mryad\\Downloads\\Input (1).xlsx"
df = pd.read_excel(file_path)

# Ensure output directory exists
output_dir = 'articles'
os.makedirs(output_dir, exist_ok=True)

# Function to extract article content
def extract_article(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            
            # Assuming the title is within <h1> tags and the article text is within <p> tags
            title = soup.find('h1').get_text(strip=True)
            paragraphs = soup.find_all('p')
            article_text = "\n".join([para.get_text(strip=True) for para in paragraphs])
            
            return title, article_text
        else:
            return None, None
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return None, None

# Iterate over each row in the dataframe
for index, row in df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    title, article_text = extract_article(url)
    if title and article_text:
        content = f"{title}\n\n{article_text}"
        
        # Save the article to a text file
        output_path = os.path.join(output_dir, f"{url_id}.txt")
        with open(output_path, 'w', encoding='utf-8') as file:
            file.write(content)
        
        print(f"Article {url_id} saved successfully.")
    else:
        print(f"Failed to extract article for URL_ID {url_id}.")

print("All articles processed.")
```
